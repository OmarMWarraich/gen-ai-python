```python
# Chat Completion
import re
from openai import OpenAI
client = OpenAI()

response = client.chat.completions.create(
    model="gpt-3.5-turbo",

    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the purpose of life?"}
    ]
)

response_content = response.choices[0].message.content
# print response_content in a more readable way
# Each line of the response is separated by a newline character, so we can replace the newline character with a space to print the response in a more readable way.

# Split content into individual sentences
sentences = re.split(r'\. ', response_content)

# Print each sentence on a new line
for sentence in sentences:
    print(sentence)
```
